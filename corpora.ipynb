```python
corpus_df = pd.read_csv(
    filename,
    sep="\t",
    header=None,
    names=["file", "token", "sentnum", "toknum", "lemma", "pos", "sem_pos", "extra", "usas"],
    dtype=str
)
```

```python
corpus_df.head()
```

фильтрация пунктуации

```python
corpus_df = corpus_df[~((corpus_df["pos"] == "Atd") | (corpus_df["usas"] == "PUNCT"))]
```

```python
corpus_df["file"].nunique()
```

```python
relevant_titles[:10]
```

```python
def make_txt(name):
    return name + ".txt"

relevant_titles = list(map(make_txt, relevant_titles))
```

```python
relevant_titles[:5]
```

оставляем только нужные файлы

```python
corpus_df = corpus_df[corpus_df["file"].isin(relevant_titles)]
```

берём только первую лемму из блока вида "abc | abc | abc"

```python
corpus_df["lemma"] = corpus_df["lemma"].str.split("|").str[0].str.strip()
```

```python
corpus_df.head()
```

```python
corpus_df["lemma"] = corpus_df["lemma"].fillna("").astype(str)
```

```python
grouped = corpus_df.groupby("file")["lemma"].apply(lambda x: " ".join(x))
```

```python
len(grouped)
```

```python
OUTPUT = "/Users/daria/Documents/welsh/d.2020.0119878310/CorCenCCcorpus/lemmatized_corpus.txt"
with open(OUTPUT, "w", encoding="utf-8") as out:
    for fname in relevant_titles:  # сохраняем исходный порядок
        text = grouped.get(fname, "")
        out.write(text + "\n")
```

```python
with open(OUTPUT, "r", encoding="utf-8") as f:
    letstest = f.readlines()
```
